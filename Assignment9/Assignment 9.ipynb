```python
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch.nn.functional as F
```

```python
model_name = "cardiffnlp/twitter-xlm-roberta-base-sentiment"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
```

```python
# Multilingual Indian language sentences

sentences = [
    # Hindi
    "यह फिल्म बहुत अच्छी है।",

    # Marathi
    "ही सेवा खूप छान आहे.",
    "मला हा अनुभव अजिबात आवडला नाही.",

    # Tamil
    "இந்த படம் மிகவும் அருமையானது.",
    "இந்த சேவை மிகவும் மோசமாக உள்ளது."
]
```

```python
labels = ['Negative', 'Neutral', 'Positive']

for sentence in sentences:
    inputs = tokenizer(sentence, return_tensors="pt", truncation=True, padding=True)

    outputs = model(**inputs)
    probs = F.softmax(outputs.logits, dim=1)

    predicted_class = torch.argmax(probs).item()

    print("Sentence:", sentence)
    print("Predicted Sentiment:", labels[predicted_class])
    print("Confidence:", round(probs[0][predicted_class].item(), 4))
    print("-" * 50)
```
